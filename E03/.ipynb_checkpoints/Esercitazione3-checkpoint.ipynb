# Plain vanilla option pricing 

## Black-Scholes analytic solution:

```python
from scipy.special import erf
from math import log, exp, sqrt

# data
S0 = 100.
K = 100.
T = 1.
r = 0.1
sigma = 0.25
t=0

def N(x):
    return 0.5 * (1. + erf(x / sqrt(2.)))

def black_scholes(S0, K, T, r, sigma):
    d1 = 1./(sigma * sqrt(T)) * (log(S0 / K) + (r + (sigma**2) / 2.) * T)
    d2 = d1 - sigma * sqrt(T)
    C = S0 * N(d1) - K * exp(-r * T) * N(d2)
    P = S0 *(N(d1) - 1.) - K * exp(-r * T) * (N(d2)-1.)
    return C, P

if __name__ == "__main__":
    call, put = black_scholes(S0, K, T, r, sigma)
    print(" == BLACK-SCHOLES ==")
    print("call: ", call)
    print("put: ", put)
```

## Exercise 03.1:

Compute at time $t=0$ **via Monte Carlo** the European call-option price, $C[S(0),0]$, and put-option price, $P[S(0),0]$:
1. By sampling **directly** the final asset price $S(T)$ for a $GBM(r,\sigma^2)$
2. By sampling the discretized $GBM(r,\sigma^2)$ path of the asset price

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

x, f, error = np.loadtxt("Data/dirC.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
ax.errorbar(x,f,yerr=error, label='direct')
ax.plot((0,99), (call,call), c='r', label='Black-Scholes solution')


x, f, error = np.loadtxt("Data/disC.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
ax.errorbar(x,f,yerr=error, label='discretized')

ax.set_xlabel('#blocks')
ax.set_title("Call")
ax.legend()
ax.grid(True)


x, f, error = np.loadtxt("Data/dirP.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
ax2.errorbar(x,f,yerr=error, label='direct')
ax2.plot((0,99), (put,put), c='r', label='Black-Scholes solution')


x, f, error = np.loadtxt("Data/disP.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
ax2.errorbar(x,f,yerr=error, label='discretized')

ax2.set_title('Put')
ax2.set_xlabel('#blocks')
ax2.legend();
ax2.grid(True)

plt.tight_layout()
plt.show()
```

All'aumentare del numero di blocchi entrambe Put e Call option, sia nel caso discreto che diretto, convergono alla soluzione analitica di Black-Scholes.
