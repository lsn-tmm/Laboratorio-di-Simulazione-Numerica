# The Traveling Salesman Problem


## Simulated Annealing 

Per l'implementazione di questo codice è stata mantenuta l'intera struttura del codice presentato nella precedente esercitazione. 
Sono state apportate le seguenti modifiche alla classe *Population*:
1. Aggiunta del membro privato temperatura $T$;
2. Aggiunta di un metodo *Boltzmann*
3. Il metodo *NewGeneration* è stato modificato per riprodurre il *Simulated Annealing* algorithm

Partendo da una temperatura fissata $T = 1$ e $N = 2 \cdot 10^4$ steps totali, dopo un numero di passi pari a $n = \frac{1}{100}N$ la temperatura viene ridotta di $\frac{1}{5}T$.

### Circumference


```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

fig = plt.figure(figsize=(14,10))
ax = fig.add_subplot(221)
ax2 = fig.add_subplot(222)

f = np.loadtxt("Es10_1/Data/cr_bestvalue.dat", usecols=(0), delimiter=' ', unpack='true')
x = [*range(len(f))]
ax.plot(x,f)

ax.set_ylim((5,35))
ax.set_title('(SA) - L1')
ax.set_xlabel('N - steps')
ax.set_ylabel('L1 - distance')
ax.grid(True)

f2 = np.loadtxt("Es10_1/GA/cr_bestvalue.dat", usecols=(0), delimiter=' ', unpack='true')
x = [*range(len(f2))]
ax2.plot(x,f2)

ax2.set_ylim((5,35))
ax2.set_title('(GA) - Best L1 each generation')
ax2.set_xlabel('N - generation')
ax2.set_ylabel('L1 - distance')
ax2.grid(True)

plt.show()

print(f'Simulated annealing: {float(f[-1:])} \t Genetic algorithm: {float(f2[-1:])}')
```

Entrambi gli algoritmi convergono alla stessa soluzione che minimizza la funzione costo $L^{(1)}$. Sotto possiamo infatti vedere che il disegno del miglior percorso sia lo stesso per *Simulated Annealig* o *Genetic Algorithm*.

```python
from math import cos, sin

fig = plt.figure(figsize=(14,14))
ax = fig.add_subplot(221)
ax2 = fig.add_subplot(222)


x = np.linspace(-1.0, 1.0, 100)
y = np.linspace(-1.0, 1.0, 100)
X, Y = np.meshgrid(x,y)
F = X**2 + Y**2 - 1
ax.contour(X,Y,F,[0])
ax2.contour(X,Y,F,[0])

_x = []
_y = []
rho,theta = np.loadtxt("Es10_1/Data/cr_bestpath.dat", usecols=(0,1), delimiter=' ', unpack='true')
for i in range(len(rho)):
    _x.append(rho[i] * cos(theta[i]))
    _y.append(rho[i] * sin(theta[i]))
    
_x2 = []
_y2 = []
rho,theta = np.loadtxt("Es10_1/GA/cr_bestpath.dat", usecols=(0,1), delimiter=' ', unpack='true')
for i in range(len(rho)):
    _x2.append(rho[i] * cos(theta[i]))
    _y2.append(rho[i] * sin(theta[i]))

ax.plot(_x,_y,'o-')

ax.set_title('(SA) - Draw best path')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_xlim((-1.5,1.5))
ax.set_ylim((-1.5,1.5))
ax.grid(True)

ax2.plot(_x2,_y2,'o-')

ax2.set_title('(GA) - Draw best path')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_xlim((-1.5,1.5))
ax2.set_ylim((-1.5,1.5))
ax2.grid(True)
plt.show()
```

### Square


```python
fig = plt.figure(figsize=(14,10))
ax = fig.add_subplot(221)
ax2 = fig.add_subplot(222)

f = np.loadtxt("Es10_1/Data/sq_bestvalue.dat", usecols=(0), delimiter=' ', unpack='true')
x = [*range(len(f))]
ax.plot(x,f)

ax.set_ylim((4,14))
ax.set_title('(SA) - L1')
ax.set_xlabel('N - steps')
ax.set_ylabel('L1 - distance')
ax.grid(True)

f2 = np.loadtxt("Es10_1/GA/sq_bestvalue.dat", usecols=(0), delimiter=' ', unpack='true')
x = [*range(len(f2))]
ax2.plot(x,f2)

ax2.set_ylim((4,14))
ax2.set_title('(GA) - Best L1 each generation')
ax2.set_xlabel('N - generation')
ax2.set_ylabel('L1 - distance')
ax2.grid(True)

plt.show()

print(f'Simulated annealing: {float(f[-1:])} \t Genetic algorithm: {float(f2[-1:])}')
```

In questo caso possiamo vedere come il *Simulated Annealing* non sia riuscito a replicare le prestazione del *Genetic Algorithm* trovando un valore leggermente superiore per $L^{(1)}$.
Sotto vengono raffigurati i percorsi migliori ottenuti da entrambi gli algoritmi.

```python
fig = plt.figure(figsize=(14,14))
ax = fig.add_subplot(221)
ax2 = fig.add_subplot(222)

ax.plot([0,1],[1,1], color='purple')
ax.plot([0,1],[0,0], color='purple')
ax.plot([0,0],[0,1], color='purple')
ax.plot([1,1],[0,1], color='purple')

ax2.plot([0,1],[1,1], color='purple')
ax2.plot([0,1],[0,0], color='purple')
ax2.plot([0,0],[0,1], color='purple')
ax2.plot([1,1],[0,1], color='purple')

x,y = np.loadtxt("Es10_1/Data/sq_bestpath.dat", usecols=(0,1), delimiter=' ', unpack='true')

ax.plot(x,y,'o-')

ax.set_title('(SA) - Draw best path')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_xlim((-0.5,1.5))
ax.set_ylim((-0.5,1.5))
ax.grid(True)

x,y = np.loadtxt("Es10_1/GA/sq_bestpath.dat", usecols=(0,1), delimiter=' ', unpack='true')

ax2.plot(x,y,'o-')

ax2.set_title('(GA) - Draw best path')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_xlim((-0.5,1.5))
ax2.set_ylim((-0.5,1.5))
ax2.grid(True)
plt.show()
```

## Parallelized Genetic Algorithm

*Genetic Search* con 4 **Genetic Algorithm paralleli**, ciascuno con una popolazione di $P = 10^4$ individui per $N = 100$ generazioni. Una migrazione avviene ogni $N_m = 10$ generazioni. Di seguito vengono confrontati i risultati con la simulazione dell'esercitazione precedente.

```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(111)

f = np.loadtxt("Es10_2/GA/sq_bestvalue.dat", usecols=(0), delimiter=' ', unpack='true')
f1 = np.loadtxt("Es10_2/Data/sq0_bestvalue.dat", usecols=(0), delimiter=' ', unpack='true')
f2 = np.loadtxt("Es10_2/Data/sq1_bestvalue.dat", usecols=(0), delimiter=' ', unpack='true')
f3 = np.loadtxt("Es10_2/Data/sq2_bestvalue.dat", usecols=(0), delimiter=' ', unpack='true')
f4 = np.loadtxt("Es10_2/Data/sq3_bestvalue.dat", usecols=(0), delimiter=' ', unpack='true')

x = [*range(len(f1))]
ax.plot(x,f, label='Genetic Algorithm', c='black')
ax.plot(x,f1, label='node 1')
ax.plot(x,f2, label='node 2')
ax.plot(x,f3, label='node 3')
ax.plot(x,f4, label='node 4')

ax.set_ylim((4,15))
ax.set_title('(PGA) - Best individual each generation')
ax.set_xlabel('N - generations')
ax.set_ylabel('L1 - distance')
ax.legend()
ax.grid(True)

plt.show()

print(f'GA: {float(f[-1:])}')
print(f'Node 1: {float(f1[-1:])} \t Node 2: {float(f2[-1:])} \t Node 3: {float(f3[-1:])} \t Node 4: {float(f4[-1:])}')
```

Possiamo vedere come il PGA migliori le prestazioni del singolo GA. Inoltre ciascuno dei quattro nodi utilizzati, anche se ad $N$ generazioni distinte, raggiungono tutti il medesimo minimo. Sotto possiamo vedere infatti rappresentato il medesimo percorso per tutti i nodi.
