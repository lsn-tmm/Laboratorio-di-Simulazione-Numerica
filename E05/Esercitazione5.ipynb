# Metropolis

Prima della presa dati di ogni caso studiato il sistema viene lasciato equilibrare con $M_{eq} = 10^5$ steps. Ciascuna simulazione Ã¨ poi realizzata con $M = 10^6$ steps totali e data blocking con $N = 100$ blocchi di $L = 10^4$ steps ciascuno.

## Wave function: $\Psi_{100}$

$$
\Psi_{1,0,0}(r,\theta,\phi)=
\frac{a_0^{-3/2}}{\sqrt{\pi}} e^{-r/a_0} 
$$

### Start from the origin $(0,0,0)$

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(221)
ax1 = fig.add_subplot(222, projection='3d')
ax2 = fig.add_subplot(223)
ax3 = fig.add_subplot(224, projection='3d')

print('Psi_100')

x, f, error = np.loadtxt("Data/psi100_uni.0", usecols=(0,1,2), delimiter=' ', unpack='true')
ax.errorbar(x,f,yerr=error)
ax.plot((0,99),(1.5,1.5), c='r', label='$<r> = 3/2 a_0$')
ax.set_xlabel('# blocks')
ax.set_ylabel('<r>')
ax.set_title("Uniform transition probability")
ax.set_ylim(1.45,1.6)
ax.grid(True)
ax.legend()

X,Y,Z = np.loadtxt("Scatter/psi100_uni.0", usecols=(0,1,2), delimiter=' ', unpack='true')
ax1.scatter(X, Y, Z, c=Z, marker='.')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('z')
ax1.view_init(10, 30)

x, f, error = np.loadtxt("Data/psi100_multi.0", usecols=(0,1,2), delimiter=' ', unpack='true')
ax2.errorbar(x,f,yerr=error)
ax2.plot((0,99),(1.5,1.5), c='r', label='$<r> = 3/2 a_0$')
ax2.set_xlabel('# blocks')
ax2.set_ylabel('<r>')
ax2.set_title("Multivariate normal transition probability")
ax2.set_ylim(1.45,1.6)
ax2.grid(True)
ax2.legend()

X,Y,Z = np.loadtxt("Scatter/psi100_multi.0", usecols=(0,1,2), delimiter=' ', unpack='true')
ax3.scatter(X, Y, Z, c=Z, marker='.')
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.set_zlabel('z')
ax3.view_init(10, 30)

plt.tight_layout()
plt.show()
```

L'origine sembra essere un buon punto di partenza per l'algoritmo di Metropolis in quanto si trova giÃ  all'interno della distribuzione di probabilitÃ  che vogliamo campionare. Entrambe le transition probability $T(\vec{x}|\vec{y})$ convergono esattamente al valore atteso.
