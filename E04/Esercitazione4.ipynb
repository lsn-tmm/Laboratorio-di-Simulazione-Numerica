# Molecular dynamics

## Solid phase
$\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)

### Equilibrating the thermodynamic phase

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111)

f = np.loadtxt("solid/eq_temp.dat", usecols=(0), delimiter=' ', unpack='true')
x = np.array([*range(len(f))])
ax.plot(10 * x,f)
ax.plot([0,len(f)*10],[0.8,0.8], color="red", label="T* target")
ax.set_xlabel('n step')
ax.set_ylabel('T*')
ax.set_title("Equilibrating temperature")
ax.grid(True)
ax.legend()
plt.show()
```

In seguito a successivi riscalamenti delle velocità (3), ogni volta facendo prima evolvere il sistema per un numero sufficiente di step ($5 \cdot 10^3)$, ed ad ulteriori $5 \cdot 10^3$ steps finali, si può vedere come questo oscilli intorno alla temperatura target $T^*$.

### State variables

```python
fig = plt.figure(figsize=(18,7))
ax1 = fig.add_subplot(121)

f, error = np.loadtxt("solid/ave_etot.dat", usecols=(0,1), delimiter=' ', unpack='true')
x =  [*range(len(f))]
ax1.errorbar(x,f,yerr=error, label='Total energy')

f, error = np.loadtxt("solid/ave_epot.dat", usecols=(0,1), delimiter=' ', unpack='true')
ax1.errorbar(x,f,yerr=error, label='Potential energy')

f, error = np.loadtxt("solid/ave_ekin.dat", usecols=(0,1), delimiter=' ', unpack='true')
ax1.errorbar(x,f,yerr=error, label='Kinetic energy')

ax1.set_xlabel('# block')
ax1.set_ylabel('E (J)')
ax1.set_title("Energy")
ax1.grid(True)
ax1.ticklabel_format(axis="y", style="sci", scilimits=(0,0))
ax1.legend()

ax2 = fig.add_subplot(122)

f, error = np.loadtxt("solid/ave_temp.dat", usecols=(0,1), delimiter=' ', unpack='true')
x =  [*range(len(f))]
ax2.errorbar(x,f, yerr=error)
ax2.set_xlabel('# blocks')
ax2.set_ylabel('T (K)')
ax2.set_title("Temperature")
ax2.grid(True)

plt.show()
```

Il grafico mostra la conservazione dell'energia totale del sistema, coerentemente col fatto che ci troviamo all'interno di un ensamble microcanonico.
Sono stati utilizzati $N = 100$ blocchi da $L = 100$ punti ciascuno.
