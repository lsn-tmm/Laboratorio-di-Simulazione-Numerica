# Monte Carlo NVT

## Solid phase
$\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)

### Equilibrating phase

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111)

x,f = np.loadtxt("MonteCarlo/eq/eq_epot.solid", usecols=(0,1), delimiter=' ', unpack='true')
ax.plot(x,f, c="r", label="U/N")

x,f = np.loadtxt("MonteCarlo/eq/eq_press.solid", usecols=(0,1), delimiter=' ', unpack='true')
ax.plot(x,f, c="g", label="P")

ax.legend();
ax.set_title('Solid Equilibration')
ax.set_xlabel('nstep')
ax.set_ylabel('values')
ax.grid(True)
plt.show()
```

Dopo un numero sufficiente di steps il sistema si equilibra, energia per particella $\frac{U}{N}$ e pressione $P$ oscillano attorno ad un valore medio.

### Autocorrelation

```python
import math

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(221)
ax2 = fig.add_subplot(222)

data = np.loadtxt("MonteCarlo/autocorrelation/ac_epot.solid", usecols=(0), delimiter=' ', unpack='true')
ax.errorbar([*range(len(data[:600]))],data[:600])
ax.set_title('Energy autocorrelation')
ax.set_xlabel('t')
ax.set_ylabel('$\chi(t)_{U/N}$')
ax.grid(True)

data = np.loadtxt("MonteCarlo/autocorrelation/ac_press.solid", usecols=(0), delimiter=' ', unpack='true')
ax2.errorbar([*range(len(data[:600]))],data[:600])
ax2.set_title('Pressure autocorrelation')
ax2.set_xlabel('t')
ax2.set_ylabel('$\chi(t)_{P}$')
ax2.grid(True)
plt.show()
```

L'autocorrelazione crolla dopo un numero sufficientemente grandi di punti, questo ci da indicazione su quanto ampi debbano almeno essere i blocchi per la blocking average.

### Statistical uncertainties

```python
def error(AV,AV2,n):  # Function for statistical uncertainty estimation
    if n==0:
        return 0
    else:
        return math.sqrt((AV2- AV**2)/n)
    
def data_blocking(l,r):
    L=l              # Number of throws in each block
    N=100       # Number of blocks
    x = np.arange(N)
    ave = np.zeros(N)
    av2 = np.zeros(N)
    sum_prog = 0
    su2_prog = 0

    for i in range(N):
        sum = 0
        for j in range(L):
            k = j+i*L
            sum += r[k]
        ave[i] = sum/L       # r_i 
        av2[i] = (ave[i])**2 # (r_i)^2 
    
    for i in range(N):
        sum_prog += ave[i] 
        su2_prog += av2[i]
    sum_prog/=N 
    su2_prog/=N
    err = error(sum_prog,su2_prog,N)
    return err

nstep = [10,100,250,500,1000,2500,5000]
U = np.loadtxt("MonteCarlo/correlation/correlation_epot.solid", usecols=(1), delimiter=' ', unpack='true')
P = np.loadtxt("MonteCarlo/correlation/correlation_press.solid", usecols=(1), delimiter=' ', unpack='true')

err_U = np.zeros(len(nstep))
err_P = np.zeros(len(nstep))
per_U = np.zeros(len(nstep))
per_P = np.zeros(len(nstep))

for i in range(len(nstep)):
    err_U[i] = data_blocking(nstep[i],U)
    err_P[i] = data_blocking(nstep[i],P)
    
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
    
ax.plot(nstep,err_U)
ax.set_title('Energy uncertainties')
ax.set_xlabel('nstep per block')
ax.set_ylabel('$\sigma_{U/N}$')
ax.grid(True)
    

ax2.plot(nstep,err_P)
ax2.set_title('Pressure uncertainties')
ax2.set_xlabel('nstep per block')
ax2.set_ylabel('$\sigma_{P}$')
ax2.grid(True)
    
plt.tight_layout()

plt.show()
```

Inizialmente aumentando il numero di punti per blocco l'errore migliora significaticamente, ma dopo un numero di steps per blocco sufficientemente grande l'errore sembra stabilizzarsi. Questo significa che anche aumentando il numero di punti non si riuscirebbe a migliorare significativamente l'errore.

### Simulation

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(221)
ax2 = fig.add_subplot(222)

x,f,error = np.loadtxt("MonteCarlo/solid/argon_epot.out", usecols=(0,2,3), delimiter=' ', unpack='true')
ax.errorbar(x,f, yerr=error)
ax.set_title('Solid - Potential energy per particle')
ax.set_xlabel('# blocks')
ax.set_ylabel('$U/N (J)$')
ax.grid(True)

x,f,error = np.loadtxt("MonteCarlo/solid/argon_pres.out", usecols=(0,2,3), delimiter=' ', unpack='true')
ax2.errorbar(x,f, yerr=error)
ax2.set_title('Solid - Pressure')
ax2.set_xlabel('# blocks')
ax2.set_ylabel('$P (Pa)$')
ax2.grid(True)

plt.tight_layout()

plt.show()
```

```python
fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot(111)

x,f,error = np.loadtxt("MonteCarlo/solid/argon_gave.out", usecols=(0,1,2), delimiter=' ', unpack='true')
ax.errorbar(x,f, yerr=error, label='Monte Carlo')
x,f,error = np.loadtxt("Molecular/solid/argon_gave.out", usecols=(0,1,2), delimiter=' ', unpack='true')
ax.errorbar(x,f, yerr=error, label='Molecular dynamics')
ax.set_title('Solid - Radial distribution function')
ax.set_xlabel('$r$')
ax.set_ylabel('$g(r)$')
ax.legend()

plt.show()
```

L'energia potenziale per particella e la pressione assumono un valore pressoch costante.
La distribuzione radiale $g(r)$ prodotta dalle due simulazioni si sovrappone esattamente.
