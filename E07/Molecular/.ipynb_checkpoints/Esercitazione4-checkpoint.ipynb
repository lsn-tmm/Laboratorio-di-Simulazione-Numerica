# Molecular dynamics

## Solid

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

f = np.loadtxt("solid/eq_temp.dat", usecols=(0), delimiter=' ', unpack='true')
x =  [*range(0,7*10**3,10)]
plt.plot(x,f)
plt.plot([0,7*10**3],[0.8,0.8], color="red", label="T* target")
plt.xlabel('n step')
plt.ylabel('T*')
plt.title("Equilibrating temperature")
plt.grid(True)
plt.legend()
plt.show()
```

### Energia
